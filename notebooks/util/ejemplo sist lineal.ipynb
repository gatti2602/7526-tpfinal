```python
import numpy as np
import matplotlib.pyplot as mp

%matplotlib inline
mp.rcParams['figure.figsize'] = [15, 6]
```

```python
def next_x(x,y):
    return 0.5*x+y+0.25

def next_y(x,y):
    return -0.5*x+y+0.75
```

```python
steps = int(100)
x = np.zeros(steps)
y = np.zeros(steps)

x[0], y[0] = 0., 0.

for i in range(1,steps):
    x[i] = next_x(x[i-1], y[i-1])
    y[i] = next_y(x[i-1], y[i-1])

mp.subplot(1,3,1)
mp.title("Phases")
mp.xlabel("x")
mp.ylabel("y")
mp.plot(x,y)

mp.subplot(1,3,2)
mp.title("x series")
mp.xlabel("steps")
mp.ylabel("x")
mp.plot(x)

mp.subplot(1,3,3)
mp.title("y series")
mp.xlabel("steps")
mp.ylabel("y")
mp.plot(y)
```
