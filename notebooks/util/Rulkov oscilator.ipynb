```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Rulkov oscilator x,y
a_list = np.arange(4.1, 4.9, 0.1)
tita = 0.001
beta = 0.001
steps = int(2000)

def next_x(x,y):
    return a / (1+x**2) + y

def next_y(x,y): 
    return y - tita*x - beta

def run_sim(): 
    global x, y
    x = np.zeros(steps)
    y = np.zeros(steps)
    x[0], y[0] = 0, -2.8

    for i in range(1,steps):
        x[i] = next_x(x[i-1], y[i-1])
        y[i] = next_y(x[i-1], y[i-1])
```

```python
a_list
```

```python
plt.rcParams['figure.figsize'] = [15, 6*len(a_list)]
i=0
for a in a_list:
    run_sim()
    plt.subplot(len(a_list),3,i*3+1)
    plt.title("Phases")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.plot(x,y)

    plt.subplot(len(a_list),3,i*3+2)
    plt.title("x series")
    plt.xlabel("steps")
    plt.ylabel("x")
    plt.plot(x)

    plt.subplot(len(a_list),3,i*3+3)
    plt.title("y series")
    plt.xlabel("steps")
    plt.ylabel("y")
    plt.plot(y)
    i+=1
```

```python
plt.rcParams['figure.figsize'] = [15, 6*len(a_list)]
i=0
for a in a_list:
    run_sim()
    plt.subplot(len(a_list),1,i+1)
    plt.title("x series")
    plt.xlabel("steps")
    plt.ylabel(f"x (a={a})")
    plt.plot(x)
    i+=1
```
